```python
import os
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
BACdata_train = pd.read_csv(Path('BACstock.csv'))
BACdata_train.head(10)
```

```python
# Use the open stock price column to train model
BACtraining_set = BACdata_train.iloc[:,1:2].values
print(BACtraining_set)
print(BACdata_train.shape)
```

```python
# normilize the dataset
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
scaledBAC_training_set  = scaler.fit_transform(BACtraining_set)

scaledBAC_training_set
```

```python
# create the X_train & Y_train data structures
X_train = []
y_train = []
for i in range (60,1258):
    X_train.append(scaledBAC_training_set[i-60:i,0])
    y_train.append(scaledBAC_training_set[i,0])
X_train = np.array(X_train)
y_train = np.array(y_train)

print(X_train.shape)
print(y_train.shape)
```

```python
# reshape the data
X_train = np.reshape(X_train,(X_train.shape[0], X_train.shape[1],1))

X_train.shape
```

```python
# Building the Model by Importing the Crucial Libraries and Adding Different Layers to LSTM.
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense 
from keras.layers import Dropout 
```

```python
regressor = Sequential()

regressor.add(LSTM(units = 50, return_sequences= True, input_shape = (X_train.shape[1],1)))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))

regressor.add(LSTM(units= 50))
regressor.add(Dropout(0.2))

regressor.add(Dense(units=1))






```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
regressor.fit(X_train, y_train, epochs =100, batch_size = 32)
```
